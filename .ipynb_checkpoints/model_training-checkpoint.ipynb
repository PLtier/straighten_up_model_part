```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
import tensorflow_hub as hub
import os
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import array_to_img
```

```python
tf.config.list_physical_devices()
```

```python
#Initialize needed variables

model = hub.load("https://tfhub.dev/google/movenet/singlepose/thunder/4")
movenet = model.signatures['serving_default']
image_size = 256;

samples_path = "../samples/official-samples/all_samples_sorted"
```

```python
image_dataset = keras.utils.image_dataset_from_directory(
    directory=samples_path,
    label_mode='binary',
    batch_size=None,
    shuffle=True,
    image_size=(256, 256),
    seed=42)

def translate_into_points(image):
    image = tf.expand_dims(image, axis=0)
    image = tf.cast(image, dtype=tf.int32)
    output=tf.squeeze(movenet(image)['output_0'])
    return output;


def map_into(filepath, label):
    return translate_into_points(filepath), label

keypoints_dataset=image_dataset.map(map_into)
print(keypoints_dataset)
```

```python
print(len(image_dataset))
```

```python
#Shuffling is already being done by train_test_split, so there's no need to to this later.
#Making 70/15/15
train_dataset, test_dataset = tf.keras.utils.split_dataset(keypoints_dataset, left_size=0.8)
train_dataset, validation_dataset = keras.utils.split_dataset(train_dataset, right_size=0.25)
```

```python
print(len(train_dataset))
print(len(test_dataset))
print(len(validation_dataset))
```

```python
train_dataset = train_dataset.batch(100)
test_dataset = test_dataset.batch(100)
validation_dataset = validation_dataset.batch(100)
```

```python
model = keras.Sequential(
    [
        keras.Input(shape=(17,3,)),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.5),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid'),
    ]
)

model.summary()
```

```python
earlystopping = keras.callbacks.EarlyStopping(monitor='val_binary_accuracy', 
                                              patience=20)
threshold = 0.5  

def mcc_metric(y_true, y_pred):
    predicted = tf.cast(tf.greater(y_pred, threshold), tf.float32)
    true_pos = tf.math.count_nonzero(predicted * y_true)
    true_neg = tf.math.count_nonzero((predicted - 1) * (y_true - 1))
    false_pos = tf.math.count_nonzero(predicted * (y_true - 1))
    false_neg = tf.math.count_nonzero((predicted - 1) * y_true)
    x = tf.cast((true_pos + false_pos) * (true_pos + false_neg) 
      * (true_neg + false_pos) * (true_neg + false_neg), tf.float32)
    return tf.cast((true_pos * true_neg) - (false_pos * false_neg), tf.float32) / tf.sqrt(x)

print(model.summary())

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy', mcc_metric ]
)


history = model.fit(train_dataset, 
                    epochs=50,
                    validation_data=validation_dataset,
                    callbacks=[earlystopping],        
         )
```

```python
# Visualize the training history to see whether you're overfitting.

plt.plot(history.history['binary_accuracy'])
plt.plot(history.history['val_binary_accuracy'])
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['TRAIN', 'VAL'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['Train', 'VAL'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['mcc_metric'])
plt.plot(history.history['val_mcc_metric'])
plt.title('MCC')
plt.ylabel('MCC')
plt.xlabel('epoch')
plt.legend(['Train', 'VAL'], loc='lower right')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python

y_pred = model.predict(test_dataset)

y_true = list(test_dataset.unbatch().map(lambda t, l: l).as_numpy_iterator())


y_pred_label = ["1", "0"]
y_true_label = ["1", "0"]

threshold = 0.5

y_true = tf.greater(y_true, 0.5)
y_pred = tf.greater(y_pred,0.5)


cm=confusion_matrix(y_true, y_pred, normalize=None)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

print(classification_report(y_true, y_pred))
```

```python
result = model.evaluate(test_dataset)
for metric, score in zip(model.metrics_names, result):
    print(metric, round(score, 3))
```
